```python

#------------------------------Import--------------------------------------------
import matplotlib.pyplot as plt
import numpy as np
import math
#------------------------------Input_Setting-------------------------------------
lamda=int(input("lamda input= "))
d=int(input("d input= "))  #最高次方 = d-1 所以會有d項


#------------------------------Def----------------------------------------------
def Transpose(A):
    A=np.array(A)
    AT=[[0]*n for i in range(d)]
    AT=np.array(AT)
    for i in range(len(A)):
        for j in range(len(A[0])):
            AT[j][i]=A[i][j]
    return(AT)

def LU_decomposition(A):
    n=len(A[0])
    L = np.zeros([n,n])
    U = np.zeros([n, n])
    for i in range(n):
        L[i][i]=1
        if i==0:
            U[0][0] = A[0][0]
            for j in range(1,n):
                U[0][j]=A[0][j]
                L[j][0]=A[j][0]/U[0][0]
        else:
                for j in range(i, n):#U
                    temp=0
                    for k in range(0, i):
                        temp = temp+L[i][k] * U[k][j]
                    U[i][j]=A[i][j]-temp
                for j in range(i+1, n):#L
                    temp = 0
                    for k in range(0, i ):
                        temp = temp + L[j][k] * U[k][i]
                    L[j][i] = (A[j][i] - temp)/U[i][i]
    return L , U

def LU_solution(L,U,b,d):    
    for i in range(d):
        b[i] = b[i]
        for j in range(d-i-1):
            k = i+1+j
            b[k] =  b[k] - b[i]*L[k][i]
    
    for i in range(d):
        h = d-i-1
        b[h] = b[h]/U[h][h]
        for j in range(d-i-1):
            k = h-1-j
            b[k] = b[k] - b[h]*U[k][h]
    return b
#----------------------------Load-------------------------------------------
data_x, data_y = np.loadtxt('./hw01_input.txt', delimiter=',', unpack=True) #delimiter=分隔 , unpack=轉換完才能用x,y存
n=len(data_x)

print("input data len is " ,len(data_x))

#Ax = b
#----------------------------A-----------------------------------
A=[[0]*(d) for i in range(n)]
for i in range(n):
    for j in range(d):
        exp = d-j-1
        A[i][j] = float(data_x[i])**exp
#----------------------------b------------------------------------
b = data_y
#--------------------------x = (AtA+lamda I)^-1 * At * b-----------------------
# Ata+lamda I
A_t = np.transpose(A)
A    = np.transpose(A_t)
identity = np.identity(d)
temp = A_t.dot(A)
temp = temp + identity *lamda
ATb  = A_t.dot(b)
L , U = LU_decomposition(temp)
answer = LU_solution(L,U,ATb,d)


print("The fitting line is ", end='')
for i in range(d):
    print(answer[i],"x^",end='')
    if i == d-1:
        print(d-i-1)
    else:
        print(d-i-1,"+",end='')
    
#-------------------------------LSE ERROR------------------------
predict = A.dot(answer)
LSE_ERROR = 0
for i in range(n):
    LSE_ERROR = LSE_ERROR + (predict[i]-b[i])**2
print("The LSE_ERROR is ",LSE_ERROR)

print("LSE figure")
x=np.linspace(-5,5)
y=np.linspace(0,100)
fitting_curve=0
for i in range(d):
    exp = d-i-1
    fitting_curve=fitting_curve + answer[i]*(x**exp)
plt.plot(data_x, data_y, '*', label='Data Point', color='black')
plt.plot(x,fitting_curve)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Visualization')
plt.legend()   # Show " * = Data Point in graph "
plt.show()


#------------------------Newton's method-------------------------
data_x, data_y = np.loadtxt('./hw01_input.txt', delimiter=',', unpack=True) #delimiter=分隔 , unpack=轉換完才能用x,y存
n=len(data_x)
#----------------------------A-----------------------------------
A=[[0]*(d) for i in range(n)]
for i in range(n):
    for j in range(d):
        exp = d-j-1
        A[i][j] = float(data_x[i])**exp
#----------------------------b------------------------------------
b = data_y
A        = np.array(A)
A        = A.astype('float64')
b        = b.astype('float64')
A_t      = np.transpose(A)
ATA      = A_t.dot(A)
H_f      = 2*ATA
L , U    = LU_decomposition(H_f)
coe_b    = np.random.randn(d,1).astype('float64')
flag = True
for i in range(1):
    temp = np.dot(np.dot(A_t,A),coe_b)
    temp = 2*temp
    temp1 = np.dot(A_t,b)
    temp1 = 2*temp1
    temp1 = temp1.reshape(d,1)
    grad_f = -(temp - temp1)
    delta  = LU_solution(L,U,grad_f,d)
    coe_b  =  delta +coe_b
Newton_answer = coe_b
Newton_answer = Newton_answer.reshape(1,d)

#print("The newton answer is " ,Newton_answer)
print("The fitting line is ", end='')
for i in range(d):
    print(Newton_answer[0][i],"x^",end='')
    if i == d-1:
        print(d-i-1)
    else:
        print(d-i-1,"+",end='')
Newton_answer = Newton_answer.reshape(d,1)
#-----------------------Newton"s method picture and error--------------------
A=np.array(A)
Newton_answer=np.array(Newton_answer)
predict = A.dot(Newton_answer)
Newton_ERROR = 0
for i in range(n):
    Newton_ERROR = Newton_ERROR + (predict[i]-b[i])**2
print("The Newton_ERROR is ",Newton_ERROR[0])

print("Newton figure")
x=np.linspace(-5,5)
y=np.linspace(0,100)
fitting_curve=0
for i in range(d):
    exp = d-i-1
    fitting_curve=fitting_curve + Newton_answer[i]*(x**exp)
plt.plot(data_x, data_y, '*', label='Data Point', color='black')
plt.plot(x,fitting_curve)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Visualization')
plt.legend()   # Show " * = Data Point in graph "
plt.show()
```
